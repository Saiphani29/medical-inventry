```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
numerical_cols=['Final_Cost','Final_Sales','Patient_ID','Quantity','RtnMRP','ReturnQuantity']
scaler=MinMaxScaler()
df_scaled=df.copy()
df_scaled[numerical_cols]=scaler.fit_transform(df[numerical_cols])
print(df_scaled.head())
```

```python
bins=[0,25,50,75,100]
labels=['0-25','25-50','50-75','75-100']
df['Final_Cost_Binned']=pd.cut(df['Final_Cost'],bins=bins,labels=labels)
print(df[['Final_Cost','Final_Cost_Binned']].head())
```

```python
from sklearn.feature_selection import VarianceThreshold
threshold=0.01
selector=VarianceThreshold(threshold=threshold)
numerical_cols=['Final_Cost','Final_Sales','Patient_ID','Quantity','RtnMRP']
selector.fit(df[numerical_cols])
zero_variance_indices=selector.get_support(indices=True)
zero_variance_features=[numerical_cols[idx] for idx in zero_variance_indices]
print('zero/near=Zero varience features:')
print(zero_variance_features)
```
